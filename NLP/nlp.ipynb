### Text Preprocessing

#### Tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer

sample_text = ['This is a chapter on text processing using tf','Text processing requires careful handling', 'tf!']
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sample_text)
```

```python
# view label of each token
word_dict = tokenizer.word_index
print(word_dict)
```

```python
# sequence of words in numerical form
seq = tokenizer.texts_to_sequences(sample_text)
print(seq)
```

```python
# padding sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_seq = pad_sequences(seq, padding='post')
print(padded_seq)
```
